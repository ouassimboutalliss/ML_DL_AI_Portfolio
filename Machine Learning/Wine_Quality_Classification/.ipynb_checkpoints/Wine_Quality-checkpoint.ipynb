# Wine quality

The datasets are related to red variants of the Portuguese "Vinho Verde" wine. For more details, consult the reference [Cortez et al., 2009]. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification tasks (good/bad). The classes are ordered and not balanced (e.g. there are much more normal wines than excellent or poor ones).

Input variables (based on physicochemical tests):
1 - fixed acidity
2 - volatile acidity
3 - citric acid
4 - residual sugar
5 - chlorides
6 - free sulfur dioxide
7 - total sulfur dioxide
8 - density
9 - pH
10 - sulphates
11 - alcohol
Output variable (based on sensory data):
12 - quality (score between 0 and 10)

#### Check for outliers in dataset (iqr)


```python
def ouliers_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 -1.5 * IQR
    upper_bound = Q3 +1.5 * IQR
    
    ls = df.index[(df[feature]  < lower_bound) | (df[feature] > upper_bound)]
    
    return ls
```

```python
index_list = []

for feature in dataset.columns:
    index_list.extend(ouliers_iqr(dataset, feature))

print("There are:", len(index_list), "outliers !")
```

#### Handling outliers

#####  Capping outliers: IQR outliers replace with 1st and 99th percentiles
This function replaces all values in the column that are below the 1st percentile with the 1st percentile and all values in the column that are above the 99th percentile with the 99th percentile. If the column data type is not float64 or int64, the function does nothing to the column and just copies it to the new dataframe.

```python
def cap_data(dataset):
    df = dataset.copy()
    for col in df.columns:
        print("capping the ",col)
        
        #check if the values is integer else skip 
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            percentiles = df[col].quantile([0.01,0.99]).values
            df.loc[df[col] <= percentiles[0], col] = percentiles[0]
            df.loc[df[col] >= percentiles[1], col] = percentiles[1]
        else:
            df[col]=df[col]
    return df
```

```python
iqr_data = cap_data(dataset)
```

## 6) Data preprocessing (split training/test, normalize, pca, feature engineering, ...)

#### Define X and y 

```python
X = iqr_data.drop('quality', axis=1)
y = iqr_data.quality
```

```python
X
```

#### Handling Imbalanced Class

```python
from imblearn.over_sampling import SMOTE
X_res, y_res = SMOTE().fit_resample(X,y)
y_res.value_counts()
```

#### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
print('X_train: ', X_train.shape)
print('y_train: ', y_train.shape)

print('X_test: ', X_test.shape)
print('y_test: ', y_test.shape)
```

```python
y_train
```

#### Create Preprocessing Pipeline

```python
# define 
StandardScaler = StandardScaler()
pca = PCA(n_components=0.90)

# define the preprocessing pipeline
preprocessing_pipeline = Pipeline([
    ('scaler', StandardScaler),
    ('pca', pca) # Reduce dimensions, retain 90% information from dataset
    # add more preprocessing steps as necessary
])


# fit and transform the preprocessing pipeline on the training data
X_train = preprocessing_pipeline.fit_transform(X_train)

# transform preprocessing pipeline on the test data
X_test = preprocessing_pipeline.transform(X_test)
```

#### After Hyperparameter Tuning (random search)

I will use random search to determine which parameters are the best for this data. 

```python

# Different ExtraTreesClassifier hyperparameters 


rf_grid = {"n_estimators": [100,200,300,700,800,1000,1200,1500],
          "criterion": ["gini", "entropy", "log_loss"],
          "max_depth": [None, 10, 20, 50, 100,150,250],
          "bootstrap": ["False", "True"],
           # Regularization techniques
          "min_samples_split": [2, 5,10,20,50,200],
          "min_samples_leaf": [1,2,10,40,50,60,200]}


                        
# Instantiate LGBMClassifier model
rs_model = RandomizedSearchCV(ExtraTreesClassifier(n_jobs=-1,
                                                random_state=42),
                                                param_distributions=rf_grid,
                                                n_iter=500, #different combinations
                                                cv=5,
                                                verbose=True)

# Fit the RandomizedSearchCV model
rs_model.fit(X_train, y_train)
```

####  Optimization AUC-ROC Threshold with f1-score

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve
from matplotlib import pyplot

# predict probabilities
yhat = rs_model.predict_proba(X_test)

# keep probabilities for the positive outcome only
yhat = yhat[:, 1]

# calculate precision and recall for different thresholds
precision, recall, thresholds = precision_recall_curve(y_test, yhat)

# convert to f1 score
f1_scores = 2 * (precision * recall) / (precision + recall)

# locate the index of the largest f1-score
ix = np.argmax(f1_scores)

# print the best threshold and corresponding f1-score
print('Best Threshold=%f, F1-Score=%.3f' % (thresholds[ix], f1_scores[ix]))

# plot precision-recall curve
no_skill = len(y_test[y_test==1]) / len(y_test)
pyplot.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
pyplot.plot(recall, precision, marker='.', label='Extra Trees Classifier')
pyplot.scatter(recall[ix], precision[ix], marker='o', color='black', label='Best')
pyplot.xlabel('Recall')
pyplot.ylabel('Precision')
pyplot.legend()
pyplot.show()
```

In this case, the optimal threshold is 0.556667.

I used the highest f1-score score to see which threshold is the best for this case.

The index of the largest f1-score is then used to find the corresponding threshold

#### Apply optimal threshold

```python
# predict probabilities on the test set
probs = rs_model.predict_proba(X_test)

# set the threshold for class 1 
threshold = 0.556667
class_1_preds = (probs[:,1] > threshold).astype(int)


# evaluate the performance of the model
print(class_1_preds)
print('\n\n')
print(y_test)
print(classification_report(y_test, class_1_preds))
print(accuracy_score(y_test, class_1_preds)*100)
print(confusion_matrix(y_test, class_1_preds))


from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, class_1_preds)
print(f"Accuracy score with threshold of {threshold}: {accuracy:.2f}")
```
